```python
# Selecciona los Pokémon cuyo 'Type 1' es igual a 'Bug' y cuyo 'Type 2' es igual a 'Poison'
bug_poison = pokemon[(pokemon['Type 1'] == 'Bug') & (pokemon['Type 2'] == 'Poison')]

# Agrupa los datos por 'Generation' y calcula la media de la columna 'Total'
grouped_data = bug_poison.groupby('Generation')['Total'].mean()

# Crea el gráfico de línea
sns.lineplot(x=grouped_data.index, y=grouped_data.values, marker='o')
plt.title('Evolución del Total para Pokémon de tipo Bug y Poison')
plt.xlabel('Generación')
plt.ylabel('Total')
plt.show()
```
